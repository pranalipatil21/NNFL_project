```python
import pandas as pd

# Load the dataset
file_path = '/home/pranali/SYprogramming/DVE/dataset/Final_End_Data1.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Add a new column for motor efficiency
# Efficiency (%) = (output power / input power) * 100
data['motor_efficiency'] = (data['output_power'] / data['input_power']) * 100

# Cap efficiency at 100%
data['motor_efficiency'] = data['motor_efficiency'].apply(lambda x: min(x, 100))

# Save the updated dataset
output_file_path = '/home/pranali/SYprogramming/DVE/dataset/Updated_End_Data_with_Efficiency1.csv'  # Replace with your desired output file path
data.to_csv(output_file_path, index=False)

print("Dataset with motor efficiency capped at 100% saved successfully.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv(r'/home/pranali/SYprogramming/DVE/dataset/Project_2.O.csv')

# Compute the correlation matrix
correlation_matrix = df.corr()

# Only plot variables with absolute correlation above a certain threshold (e.g., 0.7)
high_correlation = correlation_matrix[(correlation_matrix >= 0.7) | (correlation_matrix <= -0.7)]

# Plot heatmap of only high correlations
plt.figure(figsize=(10, 8))
sns.heatmap(high_correlation, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Filtered Correlation Heatmap (Threshold > 0.7)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV dataset
df = pd.read_csv(r'/home/pranali/SYprogramming/DVE/dataset/Project_2.O.csv')

# List of strong positive and negative correlations to plot
strong_correlations = [
    ('Voltage', 'Pack SoC'), 
    ('Voltage', 'Temperature High'),
    ('Pack SoC', 'Temperature High'),
    ('Temperature High', 'Motor_temp'),
    ('Motor_temp', 'Controller_temp'),
    ('PC', 'Torque'),
    ('PC', 'RPM'),
    ('Torque', 'RPM'),
    ('Torque', 'Controller_temp'),
    ('input_power', 'output_power'),
    ('input_power', 'speed'),
    ('output_power', 'speed'),
    # Negative correlations
    ('Voltage', 'Temperature High'),
    ('Voltage', 'Motor_temp'),
    ('Pack SoC', 'Temperature High'),
    ('Pack SoC', 'Motor_temp')
]

# Set up the plotting area
plt.figure(figsize=(15, 20))

# Plot each pair of correlated variables using scatter plots with regression lines
for i, (param1, param2) in enumerate(strong_correlations):
    plt.subplot(6, 3, i+1)
    sns.regplot(x=df[param1], y=df[param2], scatter_kws={'s':5}, line_kws={'color':'red'})
    plt.title(f'{param1} vs {param2}')
    plt.xlabel(param1)
    plt.ylabel(param2)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV dataset
df = pd.read_csv(r'/home/pranali/SYprogramming/DVE/dataset/Project_2.O.csv')

# List of strong positive and negative correlations to plot
strong_correlations = [
    ('Voltage', 'Pack SoC'), 
    ('Voltage', 'Temperature High'),
    ('Pack SoC', 'Temperature High'),
    ('Temperature High', 'Motor_temp'),
    ('Motor_temp', 'Controller_temp'),
    ('PC', 'Torque'),
    ('PC', 'RPM'),
    ('Torque', 'RPM'),
    ('Torque', 'Controller_temp'),
    ('input_power', 'output_power'),
    ('input_power', 'speed'),
    ('output_power', 'speed'),
    # Negative correlations
    ('Voltage', 'Temperature High'),
    ('Voltage', 'Motor_temp'),
    ('Pack SoC', 'Temperature High'),
    ('Pack SoC', 'Motor_temp')
]

# Sample the data to reduce the number of points plotted
df_sampled = df.sample(frac=0.2, random_state=42)  # Using 20% of the data for clarity

# Set up the plotting area
plt.figure(figsize=(15, 20))

# Plot each pair of correlated variables using scatter plots with regression lines
for i, (param1, param2) in enumerate(strong_correlations):
    plt.subplot(6, 3, i+1)
    sns.regplot(
        x=df_sampled[param1], 
        y=df_sampled[param2], 
        scatter_kws={'s':10, 'alpha':0.3, 'color':'blue'},  # smaller points with transparency
        line_kws={'color':'red'}
    )
    plt.title(f'{param1} vs {param2}', fontsize=10)
    plt.xlabel(param1)
    plt.ylabel(param2)
    plt.grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV dataset
df = pd.read_csv(r'/home/pranali/SYprogramming/DVE/dataset/Project_2.O.csv')

# List of strong positive and negative correlations to plot
strong_correlations = [
    ('Voltage', 'Pack SoC'), 
    ('Voltage', 'Temperature High'),
    ('Pack SoC', 'Temperature High'),
    ('Temperature High', 'Motor_temp'),
    ('Motor_temp', 'Controller_temp'),
    ('PC', 'Torque'),
    ('PC', 'RPM'),
    ('Torque', 'RPM'),
    ('Torque', 'Controller_temp'),
    ('input_power', 'output_power'),
    ('input_power', 'speed'),
    ('output_power', 'speed'),
    # Negative correlations
    ('Voltage', 'Temperature High'),
    ('Voltage', 'Motor_temp'),
    ('Pack SoC', 'Temperature High'),
    ('Pack SoC', 'Motor_temp')
]

# Set up the plotting area
plt.figure(figsize=(15, 20))

# Plot bar charts for each pair of correlated variables
for i, (param1, param2) in enumerate(strong_correlations):
    plt.subplot(6, 3, i+1)
    
    # Calculate means for bar plots
    means = df[[param1, param2]].mean()
    
    # Create a bar plot
    sns.barplot(x=means.index, y=means.values, palette="viridis")
    
    plt.title(f'{param1} vs {param2} (Average)', fontsize=10)
    plt.xlabel(param1)
    plt.ylabel(param2)
    plt.grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
